```python
def inicializacao():
  op = 0
  while op < 1 or op > 4:
     try:
       op = int(input("Seleciona uma opção:\n1-Inserir Produto \n2-Reiniciar escolha de Produtos \n3-Cadastro para Entrega \n4-Finalizar a compra\n"))
       print("-"*60)
     except:
        print("Opção inválida.")      
        print("-"*60,"\n")
  
  if op == 1:
    opcao1()
  elif op == 2:
    opcao2()
  elif op == 3:
    opcao3()
  else:
    opcao4()

         
```

```python
def opcao1():
  valor_compra = 0
  valor_temporario = 0

  print("-"*60)

  while True:  
   for i in range(len(nomes_produtos)):
     print(f"{i+1}-{nomes_produtos[i]} = R${preco_produtos[i]}")
  
   num = 0

   while (num < 1 or num > 10):
     try:
       print("Para sair do Cardápio digite 20.")
       num = int(input("Digite o codigo do produto(1-10):\n"))
       if num == 20:
         inicializacao()
       break
     except:
       print("Codigo invalido.") 
       continue
  
   print("-"*60)  
   qtd = 0
   nome = ""
   indice = num-1
   
   for cont in range(len(nomes_produtos)):
     if cont == (indice):
       nome = nomes_produtos[cont]

   while qtd < 1 or qtd > 11:
     try:
       qtd = int(input(f"Digite a quantidade de porções de {nome}:\n"))
       break
     except:
       print("Codigo invalido.") 
      
   for i in range(len(preco_produtos)):
     if i == (indice):
       valor_temporario = (preco_produtos[indice])*qtd
       
   listaValoresPedido.append(valor_temporario)
   listaProdutosCodigos.append(num)
   listaProdutos.append((nome)+":"+str(qtd)+" unidade(s)")
      
   print("-"*60)
   saida = exibirProdutos()
   print(f"Seu Pedido (Codigo-Nome do Produto-Quantidade):\n{saida}")
   valor_compra += valor_temporario 
   print(f"Valor total:R${valor_compra}")
   print("-"*60)
   
```
